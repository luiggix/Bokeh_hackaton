```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd 
```

```python
TFR = pd.read_csv('Datos/UNdata_Export_20230621_205538168.zip')
```

```python
TFR
```

```python
TFR_group = TFR.groupby('Country or Area')
```

```python
def plot_country(ax, label, color):
    maxim = 2021
    country = TFR_group.get_group(label)
    x = country['Year(s)'][country['Year(s)']>=maxim-1]
    y = country['Value'][country['Year(s)']>=maxim-1]
    ax.plot(x, y, c=color, ls = '--', lw = 0.75)
    
    x = country['Year(s)'][country['Year(s)']<maxim]
    y = country['Value'][country['Year(s)']<maxim]
    ax.plot(x, y, c=color, ls = '-', lw = 2.0, label=label)
```

```python
fig = plt.figure(figsize=(10,7))
ax = plt.gca()
TFR.plot(x='Year(s)', y='Value', color='silver', rot = 70, xlabel='', lw = 0.95, ax = ax, label='', legend=False)
ax.plot([1950, 2101],[2.1,2.1], c='k', ls = '-', lw=0.75)
ax.annotate('Nivel de \n reemplazo \n promedio = 2.1', 
             xy=(2090, 2.095), xytext=(2090, 4.0),
             bbox=dict(boxstyle='round', facecolor='whitesmoke', edgecolor='gray', alpha=0.75, linewidth=0.5),
             arrowprops=dict(arrowstyle='->', facecolor='dimgray', edgecolor='dimgray'),
             fontsize=8, color='black', horizontalalignment='center')

plot_country(ax, 'Mexico', 'forestgreen')
plot_country(ax, 'Niger', 'darkred')
plot_country(ax, 'Japan', 'mediumblue')
plot_country(ax, 'World', 'black')

ax.set_title('Promedio de nÃºmero de hijos por mujer', loc='left', color='dimgray', fontsize=10)
ax.set_title('fuente: http://data.un.org', loc='right', color='dimgray', fontstyle='italic', fontsize=10)
plt.suptitle('EvoluciÃ³n del TFR (Total Fertility Rate)', y = 0.96, color='black', fontsize=14)
ax.spines[['top', 'bottom', 'left', 'right']].set_visible(False)
ax.grid(lw=0.5, color='dimgray')

ax.legend()
plt.show()
```
